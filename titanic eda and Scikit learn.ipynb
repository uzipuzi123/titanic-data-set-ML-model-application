```python
import pandas as pd
import numpy as np 
df=pd.read_csv("C:\\Users\\ASUS\\titanic.csv")
dummy1 = pd.get_dummies(df['Sex'])
dummy1['male'] = dummy1['male'].apply(lambda x: 1 if x == 1 else 0)
dummy1['female'] = dummy1['female'].apply(lambda x: 1 if x == 1 else 0)
df=pd.concat([df,dummy1],axis=1)
df
df=df.drop(["Sex"],axis=1)
```

```python
df.head()
```

```python
avg_age=df.Age.mean()
avg_age
df["Age"].replace(np.nan,avg_age,inplace=True)

df
dummy2=pd.get_dummies(df["Embarked"])

dummy2['C'] = dummy2['C'].apply(lambda x: 1 if x == True else 0)
dummy2['Q'] = dummy2['Q'].apply(lambda x: 1 if x == True else 0)
dummy2['S'] = dummy2['S'].apply(lambda x: 1 if x == True else 0)

dummy2
df.drop(['Embarked'],axis=1,inplace=True)
df=pd.concat([df,dummy2],axis=1)
#df
```

```python
df.drop(['Cabin'],axis=1,inplace=True)
df
```

```python

from sklearn.model_selection import train_test_split
from sklearn.svm import LinearSVC #support vector classifier
X=df.drop(["Survived","Name","Ticket"],axis=1)
y=df.Survived
```

```python
X_train,X_test, y_train, y_test= train_test_split(X,y, test_size=0.30, random_state =42)
```

```python
X_train.shape, X_test.shape
```

```python
model = LinearSVC()
model.fit(X_train,y_train)
```

```python
from sklearn.metrics import accuracy_score

y_est =  model.predict(X_test)
accuracy_score(y_test,y_est),model.intercept_,model.coef_
```

```python
from sklearn.ensemble import RandomForestClassifier
model1 = RandomForestClassifier()
model1.fit(X_train,y_train)
y_est =  model1.predict(X_test)
accuracy_score(y_test,y_est)
```
